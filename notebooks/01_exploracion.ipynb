# FASE 1 - Exploración de bases de datos
En esta sección exploraremos al menos 3 tipos diferentes de bases de datos

# BASE DE DATOS #1
Air Pollution Image Dataset from India and Nepal
- __Descripción__: Este dataset combina imágenes fotográficas del ambiente con metadatos de calidad del aire recolectados en India y Nepal. Cada imagen está etiquetada con su clase de AQI (AQI_Class) y acompañada de mediciones de contaminantes como PM2.5, PM10, O3, CO, SO2 y NO2, permitiendo análisis
  tanto visuales como tabulares.
- __Tipo__: Imágenes y csv
- __Fuente__: Kaggle - adarshrouniyar
- __Link de fuente__: https://www.kaggle.com/datasets/adarshrouniyar/air-pollution-image-dataset-from-india-and-nepal
- __Tipo de fuente__: Secundaria
- __Registros__: 12,240 imágenes con metadatos
- __Atributos__: Location, Filename, Year, Month, Day, Hour, AQI, PM2.5, PM10, O3, CO, SO2, NO2, and AQI_Class.
- __Tipo de datos__: enteros (Year, Month, Day, Hour), decimales (AQI, PM2.5, PM10, O3, CO, SO2, NO2), cadenas de texto (Location, Filename, AQI_Class).
- __Tamaño__: 667 MB
- __Nivel de documentación__: Alta - tiene descripción de cada columna y contexto del dataset en Kaggle
- __Posibles aplicaciones__: Clasificación de niveles de contaminación por imagen,
predicción de AQI a partir de variables atmosféricas, detección visual de smog

# BASE DE DATOS #2
Mental Health Corpus
 - __Descripción__: Este dataset contiene una colección de textos relacionados con personas que padecen ansiedad, depresión y otros problemas de salud mental, extraídos de foros y comunidades en línea. Consta de dos columnas: una con el texto del comentario y otra con una etiqueta binaria que indica si el texto refleja una condición de salud mental (1) o no (0).
 - __Tipo__: Texto (NLP) - archivo CSV con texto libre
 - __Fuente__: Kaggle - reihanenamdari
 - __Link de fuente__: https://www.kaggle.com/datasets/reihanenamdari/mental-health-corpus
 - __Tipo de fuente__: Secundaria
 - __Registros__: 27,972 oraciones etiquetadas
 - __Atributos__: text, label
 - __Tamaño__: 13.11 MB
 - __Nivel de documentación__: Media - tiene descripción general de los atributos
 - __Posibles aplicaciones__:  Clasificación de texto, análisis de opiniones, análisis del lenguaje relacionado con la salud mental y detección de lenguaje tóxico.

# BASE DE DATOS #3  (SELECCIONADA)
Calidad del aire en Colombia
 - __Descripción__: Este dataset recopila mediciones de calidad del aire registradas por estaciones meteorológicas distribuidas en todo el territorio colombiano. Incluye variables de concentración de contaminantes, información geográfica de cada estación y datos temporales
  de cada medición.
 - __Tipo__: Tabular
 - __Fuente__: Datos Abiertos Colombia - IDEAM
 - __Link de fuente__: https://www.datos.gov.co/d/g4t8-zkc3
 - __Tipo de fuente__: Secundaria
 - __Registros__: 29,683 (descarga parcial del dataset oficial que contiene 32.8M registros)
 - __Atributos__: ESTACION_ID, NOMBRE_FGDA, NOMBRE_EST, MSFL_CODE, MED_CONCENTRACION_ESTANDAR, MED_FECHA_INICIO, MED_FECHA_FINAL, LATITUD, LONGITUD, ALTITUD, NOMBRE_UNIDAD, SIGLA_UNIDAD, DURACION, CODIGO_DEPARTAMENTO, DEPARTAMENTO, CODIGO_MUNICIPIO, MUNICIPIO, TIPO_ESTACION
 - __Tamaño__: 10.381 MB
 - __Nivel de documentación__:  Alta - tiene descripción general con detalle de cada atributo
 - __Posibles aplicaciones__:
    - Análisis de tendencias de contaminación del aire por departamento y municipio
    - Identificación de zonas críticas con mayor concentración de contaminantes
    - Predicción de niveles de contaminación según ubicación geográfica y altitud
    - Correlación entre tipo de estación y niveles de contaminación registrados
    - Apoyo a políticas públicas de salud ambiental en Colombia

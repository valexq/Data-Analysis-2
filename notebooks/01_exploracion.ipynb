# FASE 1 - Exploración de bases de datos
En esta sección exploraremos al menos 3 tipos diferentes de bases de datos

# BASE DE DATOS #1
Air Pollution Image Dataset from India and Nepal
- Tipo: Imágenes y csv
- Fuente: Kaggle - https://www.kaggle.com/datasets/adarshrouniyar/air-pollution-image-dataset-from-india-and-nepal
- Tipo de fuente: Secundaria
- Registros: 12,240 imágenes con metadatos
- Atributos: Location, Filename, Year, Month, Day, Hour, AQI, PM2.5, PM10, O3, CO, SO2, NO2, and AQI_Class.
- Tamaño: 667 MB
- Nivel de documentación: Alta - tiene descripción de cada columna y contexto del dataset en Kaggle
- Posibles aplicaciones: Clasificación de niveles de contaminación por imagen,
predicción de AQI a partir de variables atmosféricas, detección visual de smog

# BASE DE DATOS #2
Mental Health Corpus
 - Tipo: Texto (NLP) - archivo CSV con texto libre
 - Fuente: https://www.kaggle.com/datasets/reihanenamdari/mental-health-corpus
 - Tipo de fuente: Secundaria
 - Registros: 27,972 oraciones etiquetadas
 - Atributos: text, label
 - Tamaño: 13.11 MB
 - Nivel de documentación: Media - tiene descripción general pero sin detalle de cada atributo
 - Posibles aplicaciones: Clasificación de texto, análisis de sentimientos, detección temprana de
 depresión y ansiedad mediante NLP
